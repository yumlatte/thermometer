```python
# Import useful libraries

import pandas as pd 
import numpy as np
from datetime import datetime 
import matplotlib.pyplot as plt
```

```python
# Import cleaned data for visualization

dataset = pd.read_csv('processed-analog-data.csv')
```

```python
# Ideal Fermentation
fermentation_xvalues = dataset['time']
fermentation_bottom_yvalues = np.full((len(dataset['time']),1),21)
fermentation_top_yvalues = np.full((len(dataset['time']),1),25)
```

```python
# Plot using matplotlib
temp_time_chart = plt.figure(figsize=(15,6))

plt.plot(dataset['time'],dataset['Temperature [C]'],ls='-', label='Recorded Data')
plt.plot(fermentation_xvalues,fermentation_bottom_yvalues, label = 'Ideal Fermentation Bottom Range', color='green')
plt.plot(fermentation_xvalues,fermentation_top_yvalues, label = 'Ideal Fermentation Top Range',color='red')

plt.title('Temperature Vs. Time Profile')
plt.xlabel('Time of Reading')
plt.ylabel('Temperature [Degrees C]')

plt.ylim(0,30)
plt.xticks(rotation=90)
plt.legend(loc='lower right')

plt.show()
```

```python
# Plot using dash
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import numpy as np
from random import randint
from itertools import count

# Initialize empty lists to store data
x_vals = []
y_vals = []

index = count()

# Function to generate random data
def generate_data():
    x = next(index)
    y = randint(0, 100)
    return x, y

# Initialize Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    dcc.Graph(id='live-update-graph'),
    dcc.Interval(
        id='interval-component',
        interval=1000,  # Update every 1 second
        n_intervals=0
    )
])

# Define callback to update the graph
@app.callback(Output('live-update-graph', 'figure'),
              [Input('interval-component', 'n_intervals')])
def update_graph(n):
    # Generate new data point
    x, y = generate_data()
    x_vals.append(x)
    y_vals.append(y)
    
    # Limit the number of data points to display
    max_len = 50
    if len(x_vals) > max_len:
        x_vals.pop(0)
        y_vals.pop(0)
    
    # Create a Plotly figure
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', name='Live Data'))
    fig.update_layout(title='Live Data Plot',
                      xaxis=dict(title='Time'),
                      yaxis=dict(title='Value'))
    
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
